### Filling in the missing data and remove NaNs

```python
#gives me all the columns with nan values, the any is saying where any row of the column as a NaN value
nan_columns = train_df1.loc[:, train_df1.isnull().any()]
#this then calls gives us the sum of all the nan values in each column
nan_columns.isnull().sum().plot(kind='bar')
```

```python
train_df1.isnull().sum()
```

```python
train_df1['LotFrontage'] = train_df1['LotFrontage'].fillna(train_df1['LotFrontage'].median())
train_df1['Alley'] = train_df1['Alley'].fillna('no_alley')
train_df1['MasVnrType'] = train_df1['MasVnrType'].fillna('no_vnr')
train_df1['MasVnrArea'] = train_df1['MasVnrArea'].fillna(0)
train_df1['BsmtQual'] = train_df1['BsmtQual'].fillna('no_basement')
train_df1['BsmtCond'] = train_df1['BsmtCond'].fillna('no_basement')
train_df1['BsmtExposure'] = train_df1['BsmtExposure'].fillna('no_basement')
train_df1['BsmtFinType1'] = train_df1['BsmtFinType1'].fillna('no_basement')
train_df1['BsmtFinType2'] = train_df1['BsmtFinType2'].fillna('no_basement')
train_df1['Electrical'] = train_df1['Electrical'].fillna('unknown')
train_df1['FireplaceQu'] = train_df1['FireplaceQu'].fillna('no_fireplace')
train_df1['GarageType'] = train_df1['GarageType'].fillna('no_garage')
train_df1['GarageYrBlt'] = train_df1['GarageYrBlt'].fillna(train_df1['GarageYrBlt'].median())
train_df1['GarageFinish'] = train_df1['GarageFinish'].fillna('no_garage')
train_df1['GarageCond'] = train_df1['GarageCond'].fillna('no_garage')
train_df1['GarageQual'] = train_df1['GarageQual'].fillna('no_garage')
train_df1['PoolQC'] = train_df1['PoolQC'].fillna('no_pool')
train_df1['Fence'] = train_df1['Fence'].fillna('no_fence')
train_df1['MiscFeature'] = train_df1['MiscFeature'].fillna('nothing')

```

### C. Cross-validated models

Using Cross-validation to evaluate the performance of the three Linear Models.

```python
from sklearn.linear_model import RidgeCV, LassoCV, ElasticNetCV
from sklearn.cross_validation import cross_val_score
```

```python
cv_results = []
```

```python
#give me three answers because earlier, it said to just leave it as default
#by having it set at default for CV is 3
cvlassoraw_pipeline = make_pipeline(StandardScaler(), LassoCV(alphas=np.logspace(-2,4,7)))
cvlassoraw_pipeline.fit(X_train, y_train)
lassoraw_cvscores = cross_val_score(cvlassoraw_pipeline, X_train, y_train)
#lasso_cvscores
lasso_cvresults_raw = {'name':'cv lasso raw', 
                'model': cvlassoraw_pipeline, 
                'dataset' : 'test',
                'preprocessing': 'raw',
                'score': lassoraw_cvscores}
cv_results.append(lasso_cvresults_raw)
```

```python
#helps to put constraints on parameters, so coefficients do not get too high
cvridge_pipeline_raw = make_pipeline(StandardScaler(), RidgeCV(alphas=np.logspace(-2,4,7)))
cvridge_pipeline_raw.fit(X_train, y_train)
ridge_cvscores_raw = cross_val_score(cvridge_pipeline_raw, X_train, y_train)
ridge_cvscores_raw
ridge_cvresults_raw = {'name':'cv ridge raw', 
                'model': cvridge_pipeline_raw, 
                'dataset' : 'test',
                'preprocessing': 'raw',
                'score': ridge_cvscores_raw}
cv_results.append(ridge_cvresults_raw)
```

```python
#helps to bring coefficients close to zero
cvelasticnet_pipeline = make_pipeline(StandardScaler(), ElasticNetCV(alphas=np.logspace(-2,4,7)))
cvelasticnet_pipeline.fit(X_train, y_train)
elasticnet_cvscores_raw = cross_val_score(cvelasticnet_pipeline, X_train, y_train)
#elasticnet_cvscores
cvelasticnet_results_raw = {'name':'cv elasticnet raw', 
                'model': cvelasticnet_pipeline, 
                'dataset' : 'test',
                'preprocessing': 'raw',
                'score': elasticnet_cvscores_raw}
cv_results.append(cvelasticnet_results_raw)
```

```python
cvelasticnet_pipeline = make_pipeline(StandardScaler(), ElasticNetCV(alphas=np.logspace(-2,4,7)))
cvelasticnet_pipeline.fit(X_train_scaled, y_train)
elasticnet_cvscores_scaled = cross_val_score(cvelasticnet_pipeline, X_train_scaled, y_train)
#elasticnet_cvscores
cvelasticnet_results_scaled = {'name':'cv elasticnet scaled', 
                'model': cvelasticnet_pipeline, 
                'dataset' : 'test',
                'preprocessing': 'scaled',
                'score': elasticnet_cvscores_scaled}
cv_results.append(cvelasticnet_results_scaled)
```

```python
cvlassoscaled_pipeline = make_pipeline(StandardScaler(), LassoCV(alphas=np.logspace(-2,4,7)))
cvlassoscaled_pipeline.fit(X_train_scaled, y_train)
lasso_cvscores_scaled = cross_val_score(cvlassoscaled_pipeline, X_train_scaled, y_train)
#elasticnet_cvscores
cvlasso_results_scaled = {'name':'cv lasso scaled', 
                'model': cvlassoscaled_pipeline, 
                'dataset' : 'test',
                'preprocessing': 'scaled',
                'score': lasso_cvscores_scaled}
cv_results.append(cvlasso_results_scaled)
```

### Interpreting coefficients

```python
#This function plots the top 10 features and the bottom 10 features that impact the selling price of a house.
def plot_coef(model, top_n = 10):
    '''
    Plots the magnitude of top and bottom n coefficients
    '''
    cols = X_train.columns
    coef = model.coef_
    zipped = list(zip(cols, coef))
    zipped.sort(key=lambda x: x[1], reverse = True)
    top_10 = pd.DataFrame(zipped).head(top_n)
    bottom_10 = pd.DataFrame(zipped).tail(top_n)
    return pd.concat([top_10, bottom_10], axis=0).plot.barh(x = 0, y = 1)
```
