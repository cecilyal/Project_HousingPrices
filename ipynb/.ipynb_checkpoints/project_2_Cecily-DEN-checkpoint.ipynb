### C. Cross-validated models

Import the Cross-Validation Models for each of the Regularized Linear Models

```python
from sklearn.linear_model import RidgeCV, LassoCV, ElasticNetCV
from sklearn.cross_validation import cross_val_score
```

Perform the cross-validation using an `np.logspace(-2,4,7)`.

```python
cv_results = list()
```

```python
np.logspace(-2,4,7)
```

```python
# kind of the baseline for cross validaiton test, to double check our previous models
cv_regression = make_pipeline(StandardScaler(), LinearRegression())
cv_regression.fit(X_train, y_train)
raw_cvscore = cross_val_score(cv_regression, X_train, y_train)
raw_cvscore
```

```python
cv_train_results = {'name':'train cv regression', 
                'model': cv_regression, 
                'dataset' : 'train',
                'preprocessing': 'cv regression',
                'score': raw_cvscore}
cv_results.append(cv_train_results)
```

```python
#give me three answers because earlier, it said to just leave it as default
#by having it set at default for CV is 3
cvlasso_pipeline = make_pipeline(StandardScaler(), LassoCV())
cvlasso_pipeline.fit(X_train, y_train)
lasso_cvscores = cross_val_score(cvlasso_pipeline, X_train, y_train)
#lasso_cvscores
lasso_cvresults = {'name':'cv lasso', 
                'model': cvlasso_pipeline, 
                'dataset' : 'train',
                'preprocessing': 'lasso',
                'score': lasso_cvscores}
cv_results.append(lasso_cvresults)
```

```python
#helps to put constraints on parameters, so coefficients do not get too high
cvridge_pipeline = make_pipeline(StandardScaler(), RidgeCV())
cvridge_pipeline.fit(X_train, y_train)
ridge_cvscores = cross_val_score(cvridge_pipeline, X_train, y_train)
ridge_cvscores
ridge_cvresults = {'name':'cv ridge', 
                'model': cvridge_pipeline, 
                'dataset' : 'train',
                'preprocessing': 'ridge',
                'score': ridge_cvscores}
cv_results.append(ridge_cvresults)
```

```python
#helps to bring coefficients close to zero
cvelasticnet_pipeline = make_pipeline(StandardScaler(), ElasticNetCV())
cvelasticnet_pipeline.fit(X_train, y_train)
elasticnet_cvscores = cross_val_score(cvelasticnet_pipeline, X_train, y_train)
#elasticnet_cvscores
cvelasticnet_results = {'name':'cv elasticnet', 
                'model': cvelasticnet_pipeline, 
                'dataset' : 'train',
                'preprocessing': 'elasticnet',
                'score': elasticnet_cvscores}
cv_results.append(cvelasticnet_results)
```

Why do regularized models perform better on your test set?

They do better better on your test set because a regression line is not scaled which can lead to give you a bad picture of how it will do with unseen data. Regularized models will also help to make sure our model does not overfit the data. 

### Interpreting coefficients

For your best model, 

* plot relevant coefficients using the `plot_coef` functoin.


```python
def plot_coef(model, top_n = 10):
    '''
    Plots the magnitude of top and bottom n coefficients
    '''
    cols = X_train.columns
    coef = model.coef_
    zipped = list(zip(cols, coef))
    zipped.sort(key=lambda x: x[1], reverse = True)
    top_10 = pd.DataFrame(zipped).head(top_n)
    bottom_10 = pd.DataFrame(zipped).tail(top_n)
    return pd.concat([top_10, bottom_10], axis=0).plot.barh(x = 0, y = 1)
```
